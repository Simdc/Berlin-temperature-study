```python
def heat_map_specific_date(dataframe, start_date, end_date, fig_size=None, rot_x=None, rot_y=None, x=None,v_min=None, v_max=None ):
    import seaborn as sns
    dataframe1=dataframe.copy(deep=True)
    dataframe1=dataframe1.loc[(dataframe1['Datetime'] > start_date) & (dataframe1['Datetime'] < end_date)]
    ylabels = dataframe1["Datetime"]
    dataframe1.drop(columns='Datetime',inplace=True)
    xlabel=[]
    for x in dataframe1.columns:
        xlabel.append(x)
        

        
    plt.figure(figsize=fig_size)
    t=sns.heatmap(dataframe1, square=True,cbar_kws={'fraction' : 0.01},cmap='OrRd',linewidth=1, vmin =v_min, vmax=v_max)
    t.set_xticklabels(labels=xlabel, rotation=rot_x)
    t.set_yticklabels(ylabels, rotation=rot_y)
    
    #dataframe1.boxplot(figsize=fig_size, rot=rot_x)
    
    return None 
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(timeseries):
    dftest = adfuller(timeseries, autolag='AIC')
    result = pd.Series(dftest[0:4], index=['Test Statistic','P-value','Lags Used','No of Observations'])
    for key,value in dftest[4].items():
        result['Critical Value (%s)'%key] = value
    return result
    
    
```

```python
from scipy import signal
def ccf_values(series1, series2):
    p = series1
    q = series2
    p = (p - np.mean(p)) / (np.std(p)*len(p))
    q = (q - np.mean(q)) / (np.std(q))  
    c = np.correlate(p, q, 'full')
    return c
    
```

```python
ccf_ielts = ccf_values(LCZ_6_night_00_40['01_Amselgrund-1_UTC_05min_ta_2022-07'],LCZ_6_night_00_40['17_Martin-Hoffmannstr_UTC_05min_ta_2022-07'])
ccf_ielts
```

```python
lags = signal.correlation_lags(len(LCZ_6_night_00_40['01_Amselgrund-1_UTC_05min_ta_2022-07']), len(LCZ_6_night_00_40['17_Martin-Hoffmannstr_UTC_05min_ta_2022-07']))
```

```python
def ccf_plot(lags, ccf):
    import matplotlib.pyplot as plt
    fig, ax =plt.subplots(figsize=(9, 6))
    ax.plot(lags, ccf)
    ax.axhline(-2/np.sqrt(23), color='red', label='5% confidence interval')
    ax.axhline(2/np.sqrt(23), color='red')
    ax.axvline(x = 0, color = 'black', lw = 1)
    ax.axhline(y = 0, color = 'black', lw = 1)
    ax.axhline(y = np.max(ccf), color = 'blue', lw = 1, linestyle='--', label = 'highest +/- correlation')
    ax.axhline(y = np.min(ccf), color = 'blue', lw = 1, linestyle='--')
    ax.set(ylim = [-1, 1])
    ax.set_title('Cross Correation Amelgrund_1 and Martin_Hofman_str', weight='bold', fontsize = 15)
    ax.set_ylabel('Correlation Coefficients', weight='bold', fontsize = 12)
    ax.set_xlabel('Time Lags', weight='bold', fontsize = 12)
    plt.legend()
```

```python
ccf_plot(lags, ccf_ielts)
```
