```python
data = pd.DataFrame()
data.insert(loc=0, column='Datetime', value=LCZ_A_night_00_40['Datetime'])
data.insert(loc=1, column='Tegel_frost', value=LCZ_A_night_00_40['26_Tegel-Forst_UTC_05min_ta_2022-07'])
data.insert(loc=2, column='Elisabeth_Aue', value=LCZ_D_night_00_40['11_Elisabeth-Aue_UTC_05min_ta_2022-07'])
data=data.dropna(axis=0)
data['Datetime'] = pd.to_datetime(data['Datetime'])
data = data.resample(rule='D', on = 'Datetime').mean()
```

```python
x = np.linspace(0,22.5)
y = m*x + c
```

```python
v = pd.DataFrame()
v.insert(loc=0, column='x', value=x)
v.insert(loc=1, column='y', value=y)
```

```python
axes=data.plot.scatter(x='Elisabeth_Aue',y='Tegel_frost')
v.plot(x='x', y='y', ax=axes, label=f'slope: {m}')
axes.set_xlabel('Elisabeth_Aue')
axes.set_ylim(0,25)
axes.set_xlim(0,25)
```
