```python
from statsmodels.tsa.stattools import adfuller
def adf_test(timeseries):
    dftest = adfuller(timeseries, autolag='AIC')
    result = pd.Series(dftest[0:4], index=['Test Statistic','P-value','Lags Used','No of Observations'])
    for key,value in dftest[4].items():
        result['Critical Value (%s)'%key] = value
    return result
    
    
```

```python
adf_table = LCZ_2_A_night.drop('Datetime', axis = 1)
adf_table.apply(adf_test, axis = 0)
```

```python
from scipy import signal
def ccf_values(series1, series2):
    p = series1
    q = series2
    p = (p - np.mean(p)) / (np.std(p) * len(p))
    q = (q - np.mean(q)) / (np.std(q))  
    c = np.correlate(p, q, 'full')
    return c
    
```

```python
ccf_ielts = ccf_values(LCZ_2_A_night['LCZ_4_average'], LCZ_2_A_night['LCZ_G_average'])
ccf_ielts
```

```python
lags = signal.correlation_lags(len(LCZ_2_A_night['LCZ_4_average']), len(LCZ_2_A_night['LCZ_G_average']))
```

```python
def ccf_plot(lags, ccf):
    import matplotlib.pyplot as plt
    fig, ax =plt.subplots(figsize=(9, 6))
    ax.plot(lags, ccf)
    ax.axhline(-2/np.sqrt(23), color='red', label='5% confidence interval')
    ax.axhline(2/np.sqrt(23), color='red')
    ax.axvline(x = 0, color = 'black', lw = 1)
    ax.axhline(y = 0, color = 'black', lw = 1)
    ax.axhline(y = np.max(ccf), color = 'blue', lw = 1, linestyle='--', label = 'highest +/- correlation')
    ax.axhline(y = np.min(ccf), color = 'blue', lw = 1, linestyle='--')
    ax.set(ylim = [-1, 1])
    ax.set_title('Cross Correation LCZ_4_average and LCZ_G_average', weight='bold', fontsize = 15)
    ax.set_ylabel('Correlation Coefficients', weight='bold', fontsize = 12)
    ax.set_xlabel('Time Lags', weight='bold', fontsize = 12)
    plt.legend()
```

```python
ccf_plot(lags, ccf_ielts)
```
