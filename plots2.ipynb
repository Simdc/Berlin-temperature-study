```python
def box_plot_specific_date(dataframe, start_date, end_date, fig_size=None, rot_x=None, x=None, y_lim=None ):
    dataframe1=dataframe.copy(deep=True)
    dataframe1=dataframe1.loc[(dataframe['Datetime'] > start_date) & (dataframe['Datetime'] < end_date)]
    dataframe1.drop(columns='Datetime',inplace=True)
    print(dataframe1.median())
    label=[]
    for x in dataframe1.columns:
        label.append(x)
        
    plt.figure(figsize=fig_size)
    t=sns.boxplot(data=dataframe1)
    if y_lim is not None:
        t.set_ylim([y_lim[0],y_lim[1]])
    
    t.set_xticklabels(labels=label, rotation=rot_x)
    #dataframe1.boxplot(figsize=fig_size, rot=rot_x)
    
    return None 
```

```python
def heat_map_specific_date(dataframe, start_date, end_date, fig_size=None, rot_x=None, rot_y=None, x=None,v_min=None, v_max=None ):
    dataframe1=dataframe.copy(deep=True)
    dataframe1=dataframe1.loc[(dataframe1['Datetime'] > start_date) & (dataframe1['Datetime'] < end_date)]
    ylabels = dataframe1["Datetime"]
    dataframe1.drop(columns='Datetime',inplace=True)
    xlabel=[]
    for x in dataframe1.columns:
        xlabel.append(x)
        

        
    plt.figure(figsize=fig_size)
    t=sns.heatmap(dataframe1, square=True,cbar_kws={'fraction' : 0.01},cmap='OrRd',linewidth=1, vmin =v_min, vmax=v_max)
    t.set_xticklabels(labels=xlabel, rotation=rot_x)
    t.set_yticklabels(ylabels, rotation=rot_y)
    
    #dataframe1.boxplot(figsize=fig_size, rot=rot_x)
    
    return None 
```
